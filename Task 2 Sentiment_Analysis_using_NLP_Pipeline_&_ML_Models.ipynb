```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
import re

def preprocess_text(text):
    text = text.lower()  # 1. lowercase

    text = re.sub(r'http\S+', '', text)  # 2. remove URLs
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # 3. remove punctuation

    words = text.split()  # 4. tokenization

    words = [word for word in words if word not in stop_words]  # 5. remove stopwords

    words = [lemmatizer.lemmatize(word) for word in words]  # 6. lemmatization

    return " ".join(words)
```

```python
df['clean_text'] = df['review'].apply(preprocess_text)
```

```python
df[['review', 'clean_text']].head()
```

**NLP Preprocessing**

Text is cleaned using lowercasing, removing punctuation, stopwords removal, and lemmatization.
Text data is preprocessed by converting to lowercase, removing URLs and punctuation, eliminating stopwords, and applying lemmatization. This helps in cleaning the text and improving model performance.

```python
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer(max_features=5000)

X_bow = bow.fit_transform(df['clean_text'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf.fit_transform(df['clean_text'])
```

```python
print(X_bow.shape)
print(X_tfidf.shape)
```

**Feature Engineering**

Text data is converted into numerical form using Bag of Words and TF-IDF techniques.

Bag of Words represents text based on word frequency, while TF-IDF assigns importance to words based on their relevance in the document.

TF-IDF performs better as it captures important words and reduces the impact of common words.

```python
from sklearn.model_selection import train_test_split

y = df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.2, random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_tfidf
```

**Train-Test Split**

The dataset is divided into training and testing sets using an 80-20 split. The training data is used to train the model, while the testing data is used to evaluate its performance.

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()

nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
print(y_pred_lr[:5])
print(y_pred_nb[:5])
print(y_pred_dt[:5])
```

**Model Training**

Three machine learning models were trained: Logistic Regression, Naive Bayes, and Decision Tree. These models learn patterns from the training data and predict sentiment on the test data.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate(y_test, y_pred, model_name):
    print(f"----- {model_name} -----")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred, average='weighted'))
    print("Recall:", recall_score(y_test, y_pred, average='weighted'))
    print("F1 Score:", f1_score(y_test, y_pred, average='weighted'))
    print("\n")
```

```python
evaluate(y_test, y_pred_lr, "Logistic Regression")
evaluate(y_test, y_pred_nb, "Naive Bayes")
evaluate(y_test, y_pred_dt, "Decision Tree")
```

**Model Evaluation**

The models are evaluated using Accuracy, Precision, Recall, and F1 Score. Among all models, Logistic Regression achieved the best performance with the highest F1 Score.
