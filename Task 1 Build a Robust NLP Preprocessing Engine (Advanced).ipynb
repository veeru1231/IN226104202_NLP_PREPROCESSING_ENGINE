**Task 1: Conceptual Understanding.**

1.Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different tokens due to case sensitivity.
"Love" may appear at the beginning of a sentence or as a proper noun, while "love" is a general word.
To avoid duplication and inconsistency, we usually convert text to lowercase.


2.What happens if stopwords are not removed?

If stopwords are not removed:
- Data becomes noisy
- Model performance may decrease
- Important words get less importance
- Processing becomes slower




3.Two scenarios where removing stopwords is harmful


1  Sentiment Analysis:
   Example: "I am not happy" → removing "not" changes meaning completely.

2  Question Answering Systems:
   Example: "What is your name?" → removing "what" makes sentence meaningless.


4.Stemming vs Lemmatization

Stemming:
- Cuts words to root form (not always meaningful)
- Example: "running" → "run"

Lemmatization:
- Converts words to proper base form using vocabulary
- Example: "better" → "good"

**Task 2: Advanced Preprocessing Function.**


```python
import re

def preprocess_text(text):
    try:
        if not isinstance(text, str) or text.strip() == "":
            return []

        # Remove URLs & emails
        text = re.sub(r"http\S+|www\S+|https\S+|\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Lowercase
        text = text.lower()

        # Handle repeated characters (soooo → soo)
        text = re.sub(r"(.)\1{2,}", r"\1\1", text)

        # Remove special characters (keep alphabets and space)
        text = re.sub(r"[^a-z\s]", "", text)

        # Remove extra spaces
        text = re.sub(r"\s+", " ", text).strip()

        tokens = text.split()

        # Remove very short tokens (except no, not)
        tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

        cleaned_sentence = " ".join(tokens)

        return tokens, cleaned_sentence

    except Exception as e:
        return []
```

**Task 3: Stress Testing**


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)
    print("-" * 50)

    results.append((sentence, tokens, clean_sentence))
```

**Task 4: Token Analytics**

```python
import numpy as np

for sentence, tokens, clean_sentence in results:
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("-" * 50)
```

Which sentence had the most noise?

"Win $$$ now!!! Limited offer!!!" had high noise due to symbols and repetition.

Which sentence retained most meaningful tokens?

"I am not happy with this" retained meaningful sentiment-related words.

**Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

**Task 6: Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Run
pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```
